```python
# dataset preparation
CoraDataset=Planetoid(root='./Cora',name='Cora')
print(CoraDataset)
print(len(CoraDataset))
```

```python
class myGNN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.GCN1=torch_geometric.nn.GCNConv(CoraDataset.num_node_features,16)
        self.GCN2=torch_geometric.nn.GCNConv(16,CoraDataset.num_classes)
        self.dropout=torch.nn.Dropout(0.4)
        self.relu=torch.nn.ReLU()
        self.training=0.1
        self.logsoftmax=torch.nn.LogSoftmax(dim=1)
    def forward(self,data):
        x,edge_index=data.x,data.edge_index
        x=self.GCN1(x,edge_index)
        x=self.relu(x)
        x=self.dropout(x)
        # x=torch.nn.Dropout(x,training=self.training)
        
        x=self.GCN2(x,edge_index)
        return self.logsoftmax(x)
```

```python
# In cora dataset,train_mask denotes training set, val_mask denotes validation set and test_mask denotes test set.

device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
GNNmodel=myGNN().to(device)
CoraData=CoraDataset[0].to(device)
optimizer=torch.optim.Adam(GNNmodel.parameters(),lr=0.01,weight_decay=5e-4)
loss_function=torch.nn.NLLLoss()
GNNmodel.train()
for epoch in range(200):
    
    # CoraData=CoraDataset.to(device)
    optimizer.zero_grad()
    pred=GNNmodel(CoraData)
    loss=loss_function(pred[CoraData.train_mask],CoraData.y[CoraData.train_mask])
    # loss=torch.nn.NLLLoss(pred[CoraData.train_mask],CoraData.y[CoraData.train_mask])
    loss.backward()
    optimizer.step()
GNNmodel.eval()
_,p=GNNmodel(CoraData).max(dim=1)
# print(p)
correct=int(p[CoraData.test_mask].eq(CoraData.y[CoraData.test_mask]).sum().item())
acc=correct/int(CoraData.test_mask.sum())
print('Accuracy:{:.4f}'.format(acc))
```

GCN model(message passing)

```python
class myGCN(MessagePassing):
    def __init__(self,in_channels,out_channels):
        # Specify the aggregation function as additive(Standard implementation of Graph Convolutional Networks)
        super().__init__(aggr='add')
        self.linear=torch.nn.Linear(in_channels,out_channels)
    def forward(self,x,edge_index):
        # Preprocessing of the adjacency matrix. Add an edge from all nodes to itself
        edge_index,_=add_self_loops(edge_index,num_nodes=x.siza(0))
        x=self.linear(x)
        # The following lines of code normalize the transformed node features
        row,col=edge_index
        # Calculate the degree of each node
        deg=degree(col,x.size(0),dtype=x.dtype)
        # Calculating the inverse of the square root of a degree
        deg_inv_sqrt=deg.pow(-0.5)
        # Calculate the normalization factor for each edge
        norm=deg_inv_sqrt[row]*deg_inv_sqrt[col]
        # use propagate method to pass the message
        return self.propagate(edge_index,x=x,norm=norm)
    def message(self,x_j,norm):
        # Calculate the information coming from neighboring nodes
        return norm.view(-1,1)*x_j
```

edge convolution

```python
class myEdgeConv(MessagePassing):
    def __init__(self, in_channels, out_channels):
        # For each node, its new feature is determined by the one with the largest difference among its neighboring nodes. This is the standard architecture of EdgeConv
        super().__init__(aggr='max')
        self.mlp = torch.nn.Sequential(
            # The features of both the source and target nodes are taken into account when calculating the difference, so the dimension of the input features is twice the dimension of the original features
            torch.nn.Linear(2*in_channels, out_channels),
            torch.nn.ReLU(),
            #  Keeping the number of input and output channels the same can be viewed as an operation that linearly transforms features in a high-dimensional feature space.
            torch.nn.Linear(out_channels, out_channels))

    def forward(self, x, edge_index):
        return self.propagate(edge_index, x=x)

    def message(self, x_i, x_j):
        # Calculate the difference between the source and target node features and stitch them together with the source node features.
        # Differences between source and target node features are computed to capture local relationships between nodes.
        # The new features of each node then depend not only on the features of its neighboring nodes, but also on the interrelationships between the neighboring nodes.
        # This approach better captures the local structural information in the graph and thus improves the performance of the model.
        tmp = torch.cat([x_i, x_j-x_i], dim=1)
        # Transformation of spliced features using multilayer perceptron
        return self.mlp(tmp)
```

dynamic edge convolution

```python
class myDynamicEdgeConv(myEdgeConv):
    def __init__(self,in_channels,out_channels,k=6):
        super().__init__()
        self.k=k
    def forward(self,x,batch=None):
        # In order to dynamically construct the neighbor relationship of each node, so as to efficiently capture the relationship between nodes in dynamic graph convolution and achieve more flexible graph convolution operations
        edge_index=knn_graph(x,self.k,batch,loop=False,flow=self.flow)
        return super().forward(x,edge_index)
```
